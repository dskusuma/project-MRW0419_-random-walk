### Penerapan Random Walk

*Random walk* sering digunakan sebagai bentuk usaha untuk menjelaskan tentang fenomena alam; seperti fisika dan kimia, *materials science*, biologi dan bidang-bidang lainnya. Berikut adalah beberapa penerapan dari *random walk*:

1. Pada bidang ekonomi finansial, hipotesis *random walk* digunakan untuk memodelkan harga *share* dan faktor-faktor lain.
2. Pada populasi genetik, *random walk* mendeskripsikan properti statistik dari *generic drift*
3. Pada komputer sains, *random walk* digunakan untuk memperkirakan ukuran dari Web.
4. Pada *image segmentation*, *random walk* digunakan untuk menentukan label yang diasosiasikan pada setiap piksel. Algoritma ini biasanya dikaitkan dengan *random walker segmentation algorithm*.

### *Random Walk* 1-Dimensi

Pada dasarnya, simulasi *random walk* 1-dimensi ini adalah simulasi di mana objek simulasi kita bergerak pada bidang berdimensi 1, misal pada garis $ x $.

Intuisi sederhana yang dapat kita gunakan:

### *Random Walk* 2 Dimensi
Kita memulai pada titik awal ( x = 0, y = 0) dan mengambil langkah acak dengan kemungkinan sebanyak sembilan, di mana setiap langkah mengikuti aturan $ \{ \Delta x , \Delta y \} \in \{ 1, 0, -1\}$ , yang menghasilkan matriks kemungkinan:




```python
# Parameter untuk proses jalan
# Jumlah dimensi yang digunakan, 2 karena bergerak di garis x dan y.
dimensi = 2 
# Jumlah step
jumlah_step = 1000 
# Pilihan step untuk menentukan bergerak ke arah mana
set_step = [ -1, 0, 1 ]
# Titik awal
origin = np.zeros(( 1, dimensi ))

print("Titik awal", origin)
```

```python
# Simulasi 2D
shape_step = ( jumlah_step, dimensi )
steps = np.random.choice( a = set_step, size = shape_step )
path = np.concatenate([ origin, steps ]).cumsum(0)

start = path[:1]
stop = path [-1:]

print("Step Shape: ", shape_step)
print("steps", steps)
```

```python
fig = plt.figure( figsize = ( 8,8 ), dpi = 100 )
ax = fig.add_subplot( 111 )
ax.scatter(path[:,0], path[:,1],c='blue',alpha=0.25,s=0.05);
ax.plot(path[:,0], path[:,1],c='blue',alpha=0.5,lw=0.25,ls='-');
ax.plot(start[:,0], start[:,1],c='red', marker='+')
ax.plot(stop[:,0], stop[:,1],c='black', marker='o')
plt.title('2D Random Walk')
plt.tight_layout(pad=0)
```
