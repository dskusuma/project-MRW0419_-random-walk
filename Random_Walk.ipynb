# Random Walk

## Pendahuluan
Random walk adalah proses mengambil langkah (berturut-turut), secara acak. Kita bisa menambahkan kondisi-kondisi tertentu sesuai dengan kasus yang diselesaikan.

Beberapa contoh kasus *Random Walk* adalah Brownian Motion of partickles, Stock Ticker movement, dan Living Cell movement
